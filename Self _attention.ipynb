```python
import tensorflow as tf
from tensorflow import keras
from keras_tuner.tuners import RandomSearch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
```

# self attention

```python
# Define the Self-Attention Layer
class SelfAttention(keras.layers.Layer):
    def __init__(self, units):
        super(SelfAttention, self).__init__()
        self.units = units

    def build(self, input_shape):
        # Create weight matrices for query, key, value
        self.query = self.add_weight(shape=(input_shape[-1], self.units),
                                     initializer="random_normal", trainable=True)
        self.key = self.add_weight(shape=(input_shape[-1], self.units),
                                   initializer="random_normal", trainable=True)
        self.value = self.add_weight(shape=(input_shape[-1], self.units),
                                     initializer="random_normal", trainable=True)

    def call(self, inputs):
        # Compute query, key, value
        q = tf.matmul(inputs, self.query)
        k = tf.matmul(inputs, self.key)
        v = tf.matmul(inputs, self.value)

        # Compute attention scores (scaled dot-product attention)
        score = tf.matmul(q, k, transpose_b=True)
        score = score / tf.math.sqrt(tf.cast(self.units, tf.float32))

        # Apply softmax to get attention weights
        attention_weights = tf.nn.softmax(score, axis=-1)

        # Get the weighted sum of the values
        context = tf.matmul(attention_weights, v)

        return context

# Define the model-building function with self-attention
def build_model(hp):
    model = keras.Sequential()

    # Add input layer
    model.add(keras.layers.InputLayer(input_shape=(X_train_scaled.shape[1],)))

    # Add the self-attention layer
    attention_units = hp.Int('attention_units', min_value=32, max_value=512, step=32)
    model.add(SelfAttention(units=attention_units))

    # Add a dense layer with hyperparameters
    model.add(keras.layers.Dense(units=hp.Int('dense_units', min_value=32, max_value=512, step=32), activation='relu'))

    # Output layer for binary classification
    model.add(keras.layers.Dense(1, activation='sigmoid'))

    # Compile the model with hyperparameters like optimizer and learning rate
    model.compile(
        optimizer=keras.optimizers.Adam(hp.Choice('learning_rate', [1e-2, 1e-3,5e-2,2e-4,5e-8, 1e-4,1e-5,1e-6,1e-7,1e-8])),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model

# Initialize the tuner
tuner = RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=7,  # Number of different models to try
    executions_per_trial=5,  # Run each model 3 times to reduce noise
    directory='my_dir',
    project_name='self_attention_tuning'
)

# Search for the best hyperparameters
tuner.search(X_train_scaled, y_train, epochs=30, batch_size=32, validation_split=0.2)

# Get the best hyperparameters
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

# Build the best model with the optimal hyperparameters
best_model = tuner.hypermodel.build(best_hps)

# Train the best model
history = best_model.fit(X_train_scaled, y_train, epochs=10, validation_split=0.2)

# Test the model on the test set
y_pred = best_model.predict(X_test_scaled)
y_pred_class = (y_pred > 0.5).astype(int)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred_class)
print(f"Test Accuracy: {accuracy * 100:.2f}%")

# Detailed classification report
print(classification_report(y_test, y_pred_class))
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Layer

class LocalAttention(Layer):
    def __init__(self, window_size):
        super(LocalAttention, self).__init__()
        self.window_size = window_size  # Size of the local attention window

    def build(self, input_shape):
        # Weights for query, key, value within the window
        self.query = self.add_weight(shape=(input_shape[-1], input_shape[-1]),
                                     initializer="random_normal", trainable=True)
        self.key = self.add_weight(shape=(input_shape[-1], input_shape[-1]),
                                   initializer="random_normal", trainable=True)
        self.value = self.add_weight(shape=(input_shape[-1], input_shape[-1]),
                                     initializer="random_normal", trainable=True)

    def call(self, inputs):
        batch_size, seq_len, features = tf.shape(inputs)[0], tf.shape(inputs)[1], tf.shape(inputs)[2]
        output = []

        for i in range(seq_len):
            # Define local window boundaries
            left = max(0, i - self.window_size // 2)
            right = min(seq_len, i + self.window_size // 2 + 1)

            # Local window inputs
            window = inputs[:, left:right, :]

            # Compute query, key, and value for local window
            q = tf.matmul(window, self.query)
            k = tf.matmul(window, self.key)
            v = tf.matmul(window, self.value)

            # Compute attention scores and apply softmax
            score = tf.matmul(q, k, transpose_b=True)
            score = score / tf.math.sqrt(tf.cast(features, tf.float32))
            attention_weights = tf.nn.softmax(score, axis=-1)

            # Weighted sum of values
            context = tf.matmul(attention_weights, v)

            # Collect the center context (attention on the current element)
            output.append(context[:, self.window_size // 2, :])

        # Stack results for each time step
        return tf.stack(output, axis=1)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Layer

class LocalAttention(Layer):
    def __init__(self, window_size):
        super(LocalAttention, self).__init__()
        self.window_size = window_size

    def build(self, input_shape):
        self.query = self.add_weight(shape=(input_shape[-1], input_shape[-1]),
                                     initializer="random_normal", trainable=True)
        self.key = self.add_weight(shape=(input_shape[-1], input_shape[-1]),
                                   initializer="random_normal", trainable=True)
        self.value = self.add_weight(shape=(input_shape[-1], input_shape[-1]),
                                     initializer="random_normal", trainable=True)

    def call(self, inputs):
        batch_size, seq_len, features = tf.shape(inputs)[0], tf.shape(inputs)[1], tf.shape(inputs)[2]

        # Define local window boundaries
        def compute_context(i):
            left = tf.maximum(0, i - self.window_size // 2)
            right = tf.minimum(seq_len, i + self.window_size // 2 + 1)
            window = inputs[:, left:right, :]

            q = tf.matmul(window, self.query)
            k = tf.matmul(window, self.key)
            v = tf.matmul(window, self.value)

            score = tf.matmul(q, k, transpose_b=True)
            score = score / tf.math.sqrt(tf.cast(features, tf.float32))
            attention_weights = tf.nn.softmax(score, axis=-1)

            context = tf.matmul(attention_weights, v)
            return context[:, self.window_size // 2, :]

        contexts = tf.map_fn(lambda i: compute_context(i), tf.range(seq_len), dtype=tf.float32)
        return tf.reduce_mean(contexts, axis=0)

    def compute_output_shape(self, input_shape):
        return (input_shape[0], input_shape[-1])

# Example model definition and hyperparameter tuning setup
def build_local_attention_model(hp):
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.InputLayer(input_shape=(X_train_scaled.shape[1], 1)))
    window_size = hp.Int('window_size', min_value=3, max_value=15, step=2)
    model.add(LocalAttention(window_size=window_size))
    model.add(tf.keras.layers.Dense(units=hp.Int('dense_units', min_value=32, max_value=512, step=32), activation='relu'))
    model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
    model.compile(
        optimizer=tf.keras.optimizers.Adam(hp.Choice('learning_rate', [1e-2, 1e-3,5e-2,2e-4,5e-8, 1e-4,1e-5,1e-6,1e-7,1e-8])),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

# Initialize and run Keras Tuner for local attention
from keras_tuner.tuners import RandomSearch

tuner_local_attention = RandomSearch(
    build_local_attention_model,
    objective='val_accuracy',
    max_trials=5,
    executions_per_trial=3,
    directory='local_attention_tuning',
    project_name='local_attention'
)

# Perform the search
tuner_local_attention.search(X_train_scaled, y_train, epochs=10, batch_size=32, validation_split=0.2)

# Retrieve best hyperparameters and build the best model
best_hps_local_attention = tuner_local_attention.get_best_hyperparameters(num_trials=1)[0]
best_model_local_attention = tuner_local_attention.hypermodel.build(best_hps_local_attention)
history_local_attention = best_model_local_attention.fit(X_train_scaled, y_train, epochs=10, validation_split=0.2)

# Evaluate the model on the test set
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    y_pred_class = (y_pred > 0.5).astype(int)
    accuracy = accuracy_score(y_test, y_pred_class)
    report = classification_report(y_test, y_pred_class)
    conf_matrix = confusion_matrix(y_test, y_pred_class)
    return accuracy, report, conf_matrix

accuracy_la, report_la, conf_matrix_la = evaluate_model(best_model_local_attention, X_test_scaled, y_test)

# Print results
print("Local-Attention Model:")
print(f"Test Accuracy: {accuracy_la * 100:.2f}%")
print(report_la)

# Plot confusion matrix
def plot_confusion_matrix(cm, title='Confusion Matrix', cmap=plt.cm.Blues):
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap=cmap, cbar=False, annot_kws={"size": 16},
                xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.title(title)
    plt.show()

plot_confusion_matrix(conf_matrix_la, title='Confusion Matrix for Local-Attention Model')
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Layer, Dropout, Dense
from tensorflow.keras.regularizers import l2
from keras_tuner.tuners import RandomSearch
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

class LocalAttention(Layer):
    def __init__(self, window_size):
        super(LocalAttention, self).__init__()
        self.window_size = window_size

    def build(self, input_shape):
        self.query = self.add_weight(shape=(input_shape[-1], input_shape[-1]),
                                     initializer="random_normal", trainable=True)
        self.key = self.add_weight(shape=(input_shape[-1], input_shape[-1]),
                                   initializer="random_normal", trainable=True)
        self.value = self.add_weight(shape=(input_shape[-1], input_shape[-1]),
                                     initializer="random_normal", trainable=True)

    def call(self, inputs):
        batch_size, seq_len, features = tf.shape(inputs)[0], tf.shape(inputs)[1], tf.shape(inputs)[2]

        def compute_context(i):
            left = tf.maximum(0, i - self.window_size // 2)
            right = tf.minimum(seq_len, i + self.window_size // 2 + 1)
            window = inputs[:, left:right, :]

            q = tf.matmul(window, self.query)
            k = tf.matmul(window, self.key)
            v = tf.matmul(window, self.value)

            score = tf.matmul(q, k, transpose_b=True)
            score = score / tf.math.sqrt(tf.cast(features, tf.float32))
            attention_weights = tf.nn.softmax(score, axis=-1)

            context = tf.matmul(attention_weights, v)
            return context[:, self.window_size // 2, :]

        contexts = tf.map_fn(lambda i: compute_context(i), tf.range(seq_len), dtype=tf.float32)
        return tf.reduce_mean(contexts, axis=0)

    def compute_output_shape(self, input_shape):
        return (input_shape[0], input_shape[-1])

# Model definition with regularization
def build_local_attention_model(hp):
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.InputLayer(input_shape=(X_train_scaled.shape[1], 1)))
    window_size = hp.Int('window_size', min_value=3, max_value=15, step=2)
    model.add(LocalAttention(window_size=window_size))

    # Add dropout and regularization
    model.add(Dropout(rate=hp.Float('dropout_rate', min_value=0.2, max_value=0.5, step=0.1)))
    model.add(Dense(units=hp.Int('dense_units', min_value=32, max_value=512, step=32),
                    activation='relu',
                    kernel_regularizer=l2(hp.Float('l2_reg', min_value=1e-5, max_value=1e-2, step=1e-5))))
    model.add(Dropout(rate=hp.Float('dropout_rate', min_value=0.2, max_value=0.5, step=0.1)))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(
        optimizer=tf.keras.optimizers.Adam(hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4])),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

# Initialize and run Keras Tuner for local attention
tuner_local_attention = RandomSearch(
    build_local_attention_model,
    objective='val_accuracy',
    max_trials=10,
    executions_per_trial=3,
    directory='local_attention_tuning',
    project_name='local_attention'
)

# Perform the search
tuner_local_attention.search(X_train_scaled, y_train, epochs=20, batch_size=32, validation_split=0.2)

# Retrieve best hyperparameters and build the best model
best_hps_local_attention = tuner_local_attention.get_best_hyperparameters(num_trials=1)[0]
best_model_local_attention = tuner_local_attention.hypermodel.build(best_hps_local_attention)
history_local_attention = best_model_local_attention.fit(X_train_scaled, y_train, epochs=20, validation_split=0.2)

# Evaluate the model on the test set
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    y_pred_class = (y_pred > 0.5).astype(int)
    accuracy = accuracy_score(y_test, y_pred_class)
    report = classification_report(y_test, y_pred_class)
    conf_matrix = confusion_matrix(y_test, y_pred_class)
    return accuracy, report, conf_matrix

accuracy_la, report_la, conf_matrix_la = evaluate_model(best_model_local_attention, X_test_scaled, y_test)

# Print results
print("Local-Attention Model:")
print(f"Test Accuracy: {accuracy_la * 100:.2f}%")
print(report_la)

# Plot confusion matrix
def plot_confusion_matrix(cm, title='Confusion Matrix', cmap=plt.cm.Blues):
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap=cmap, cbar=False, annot_kws={"size": 16},
                xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.title(title)
    plt.show()

plot_confusion_matrix(conf_matrix_la, title='Confusion Matrix for Local-Attention Model')
```
